```python
import numpy as np
import pandas as pd

def remove_outliers_iqr(data, k=1.5):
    cleaned_data = data.copy()  # Create a copy of the DataFrame
    
    for column in cleaned_data.columns:
        if column not in ['id', 'cholesterol', 'gluc','alco','cardio','gender']:  # Exclude 'id' column
            q1 = np.percentile(cleaned_data[column], 25)
            q3 = np.percentile(cleaned_data[column], 75)
            iqr = q3 - q1
            lower_bound = q1 - k * iqr
            upper_bound = q3 + k * iqr
            mask = (cleaned_data[column] >= lower_bound) & (cleaned_data[column] <= upper_bound)
            cleaned_data = cleaned_data[mask]
    
    return cleaned_data


cleaned_data = remove_outliers_iqr(data)

print("Cleaned Data:")
print(cleaned_data)
```
